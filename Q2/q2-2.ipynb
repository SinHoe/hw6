```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    
    f = open('dice.csv','r',encoding = 'ansi')
    data = csv.reader(f)
    
    dice100 = []
    dice1000 = []
    dice10000 = []
    dice100000 = []
    dice_list = [dice100,dice1000,dice10000,dice100000]
    i = 0
    for row in data:
        roll = 10 ** (i + 2)
        for num in range(1,roll):
            row[num] = int(row[num])
            dice_list[i].append(row[num])
        i += 1
    f.close()
    
    plt.rc('font', family='Batang')
    fig, axes = plt.subplots(4,1,figsize=(7, 12))
    plt.tight_layout()
    axes[0].hist(dice100)
    axes[0].set_title('100회 시행')
    axes[0].set_xlabel('주사위 눈')
    axes[0].set_ylabel('나온 횟 수')
    axes[1].hist(dice1000)
    axes[1].set_title('1000회 시행')
    axes[1].set_xlabel('주사위 눈')
    axes[1].set_ylabel('나온 횟 수')
    axes[2].hist(dice10000)
    axes[2].set_title('10000회 시행')
    axes[2].set_xlabel('주사위 눈')
    axes[2].set_ylabel('나온 횟 수')
    axes[3].hist(dice100000)     
    axes[3].set_title('100000회 시행')
    axes[3].set_xlabel('주사위 눈')
    axes[3].set_ylabel('나온 횟 수')
        
```
