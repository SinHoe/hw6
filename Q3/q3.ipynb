```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    
    f = open('Jeju.csv','r',encoding = 'ansi')
    data = csv.reader(f)
    
    m_population = []
    w_population = []
    for row in data:
        
        if row[0] == '제주특별자치도  (5000000000)':
            for i in range(3,21):
                row[i]=row[i].replace(',','')
            m_population.append(int(row[3]))
            m_population.append(int(row[7]))
            m_population.append(int(row[11]))
            m_population.append(int(row[15]))
            m_population.append(int(row[19]))
            w_population.append(int(row[4]))
            w_population.append(int(row[8]))
            w_population.append(int(row[12]))
            w_population.append(int(row[16]))
            w_population.append(int(row[20]))
    f.close()
    
    year = [2018,2019,2020,2021,2022]
    plt.rc('font', family='Batang')
    plt.xticks([2018,2019,2020,2021,2022])
    plt.ylabel('인구 수')
    plt.xlabel('년도')
    plt.title('제주도 년도별 남,여 인구 수')
    plt.plot(year,m_population,'b',label='남자')
    plt.plot(year,w_population,'r',label='여자')
    plt.legend()
    plt.show()
            
```
